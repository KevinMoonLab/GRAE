```python
from grae.models.semisupervised import semisupervised_all3
from grae.data import manifolds
import torch
import umap 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import numpy as np
import pandas as pd
from grae.data.base_dataset import DEVICE
device = DEVICE
```

```python
plabels = np.array([0.01, 0.05])
lambdas = [0, 1]
RUNS = 8
DATASETS = ["MnistDigits"]
Models = ["semisupervised_all3"]

# pandas data frame 
columns_names = ["Model", "Dataset", "Test acc", "Labeled percentage", "RUN", "lambda"]
inds = []
df = pd.DataFrame(data=None, index=inds, columns=columns_names, 
                 dtype=None, copy=False)

for DATAS in DATASETS:
    for per_labels in plabels:
        for run in range(RUNS):
            data_train = getattr(manifolds, DATAS)(perc_labels = per_labels, n_samples = 5000, subsetL = False)
            Embedding = umap.UMAP(n_neighbors=20, n_components = 5, min_dist=0.01).fit_transform(data_train.data)
            for Model in Models:
                for lamb in lambdas:
                    if lamb == 0:
                        Epo = 1000
                        Parameters = dict(epochs = Epo,
                          lam = 0,
                          batch_size = 256,
                          regression = False,
                          lr = 0.0001,
                          Embedding = Embedding
                          )
                    else:
                        Epo = 1000
                        Parameters = dict(epochs = Epo,
                          lam = lamb,
                          batch_size = 256,
                          regression = False,
                          lr = 0.0001,
                          Embedding = Embedding
                          )

                    # compute grae and latent representation 
                    m = semisupervised_all3(**Parameters)
                    m.fit(data_train)
                    y_p, *R1 = m.torch_module(data_train.data.float().to(device))
                    predictedLab = y_p.argmax(1).detach().cpu().numpy()


                    '''compute accuracy'''
                    acc = np.sum(np.equal(data_train.true_labels[data_train.test_ind], predictedLab[data_train.test_ind]))/len(data_train.true_labels[data_train.test_ind])
                    df2 = pd.DataFrame([[Model, DATAS, acc, str(per_labels*100)+ '%', run, lamb]], columns=columns_names)
                    df = df.append(df2)
```
